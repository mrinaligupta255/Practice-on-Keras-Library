```
from keras.preprocessing.image import ImageDataGenerator
from keras.datasets import mnist
import matplotlib.pyplot as plt


#load data

(x_train,y_train),(x_test,y_test)=mnist.load_data()
print(x_train.shape)
#reshaping the data to [sample][width][height][channel]

x_train=x_train.reshape(x_train.shape[0],28,28,1)
x_test=x_test.reshape(x_test.shape[0],28,28,1)


```

```
for j in range(9):
    plt.subplot(330+1+j)
    plt.imshow(x_train[j].reshape(28,28))
```

##1. Feature Standardization

```
#creating datagenerator
x_train1=x_train
generator=ImageDataGenerator(featurewise_std_normalization=True ,featurewise_center=True)
generator.fit(x_train1)
#configure batch size and retreive the one batch
for x,y in generator.flow(x_train1,y_train,batch_size=9,seed=1):
  for j in range(9):
    plt.subplot(330+1+j)
    plt.imshow(x[j].reshape(28,28),cmap=plt.get_cmap('gray'))
  
  plt.show()
  break

```

##2.ZCA Whitenting

```
#creating datagenerator
x_train1=x_train
generator=ImageDataGenerator(zca_whitening=True)
generator.fit(x_train1)
#configure batch size and retreive the one batch
for x,y in generator.flow(x_train1,y_train,batch_size=9,seed=1):
  for j in range(9):
    plt.subplot(330+1+j)
    plt.imshow(x[j].reshape(28,28),cmap=plt.get_cmap('gray'))
  
  plt.show()
  break
```

##3. Random Rotations

```
#creating datagenerator
x_train1=x_train
generator=ImageDataGenerator(rotation_range=90)
generator.fit(x_train1)
#configure batch size and retreive the one batch
for x,y in generator.flow(x_train1,y_train,batch_size=9,seed=1):
  for j in range(9):
    plt.subplot(330+1+j)
    plt.imshow(x[j].reshape(28,28),cmap=plt.get_cmap('gray'))
  
  plt.show()
  break
```

##4. Flipping the data horizontally and vertically

```
#creating datagenerator
x_train1=x_train
generator=ImageDataGenerator(horizontal_flip=True,vertical_flip=True)
generator.fit(x_train1)

#configure batch size and retreive the one batch
for x,y in generator.flow(x_train1,y_train,batch_size=9,seed=1):
  for j in range(9):
    plt.subplot(330+1+j)
    plt.imshow(x[j].reshape(28,28),cmap=plt.get_cmap('gray'))
  
  plt.show()
  break
```

##5. Random shifting the data

```
#creating datagenerator
x_train1=x_train
generator=ImageDataGenerator(width_shift_range=0.2,height_shift_range=0.4)
generator.fit(x_train1)
#configure batch size and retreive the one batch
for x,y in generator.flow(x_train1,y_train,batch_size=9,seed=1):
  for j in range(9):
    plt.subplot(330+1+j)
    plt.imshow(x[j].reshape(28,28),cmap=plt.get_cmap('gray'))
  
  plt.show()
  break
```
